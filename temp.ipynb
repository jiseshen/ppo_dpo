```python
import pandas as pd



a = "Flan-T5_DPO_LoRA_HH-RLHF - eval/logps/chosen"
b = "Flan-T5_DPO_LoRA_HH-RLHF - eval/logps/rejected"

chosen = pd.read_csv("Chosen.csv")[a].to_numpy()
rejected = pd.read_csv("Rejected.csv")[b].to_numpy()
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot([100 * i for i in range(len(chosen))],chosen, label="Chosen")
plt.plot([100 * i for i in range(len(rejected))], rejected, label="Rejected")
plt.legend()
plt.xlabel("Step")
plt.ylabel("LogP")
plt.title("Validation Log Probabilities of Chosen and Rejected Actions")
plt.show()
```
